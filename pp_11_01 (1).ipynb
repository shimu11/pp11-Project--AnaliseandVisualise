```python
df = pd.read_csv("student_data.csv")
df.head()
df.sample(5)
```

```python
print("Total number of entries in the dataset:", len(df))
```

```python
cgpa_threshold = 3.5
selected_students = df[df['cgpa'] > cgpa_threshold]
selected_students
```

```python
average_attendance = df['attendance'].mean()
print(f"The average attendance percentage is: {average_attendance:.2f}%")
```

```python
df['home_district'] = df['home_district'].fillna("unknown")
print(df)
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(df['cgpa'], df['attendance'])
plt.xlabel('CGPA')
plt.ylabel('Attendance')
plt.title('CGPA vs. Attendance')
plt.grid(True)
plt.show()
```
